PREPARACION DEL DATASET

```python
import pandas as pd
import seaborn as sns

df = sns.load_dataset("penguins")

#borrar del dataset todas las filas que tengan algún valor 'NA'.
df = df.dropna()
print(df.head().T)
```

CODIFICACION VARIABLE OBJETIVO

```python
#Convertimos la variable objetivo a valores numéricos
df.species = df.species.astype('category').cat.codes
```

DIVIDIR CONJUNTO EN ENTRENAMIENTO (80%) Y TEST (20%)

```python
from sklearn.model_selection import train_test_split
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=1, stratify=df['species'])
```

DEFINIR VARIABLES CATEGORICAS Y NUMERICAS Y CONVERTIR A LISTA D DICCIONARIOS

```python
from sklearn.feature_extraction import DictVectorizer

#Convertir a lista de diccionarios
categorical = ['island', 'sex']
numerical = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g']
train_dict = df_train_full[categorical + numerical].to_dict(orient='records')
test_dict = df_test[categorical + numerical].to_dict(orient='records')
```

CODIFICACION VARIABLES CATEGORICAS ONE-HOT

```python
#DictVectorizer para codificación de sus propiedades
dv = DictVectorizer(sparse=False)
dv.fit(train_dict) #ajustar a los datos de entrenamiento
```

```python
#Transformar el conjunto de entrenamiento a matriz
X_train = dv.transform(train_dict)
X_test = dv.transform(test_dict)
```

```python
#nombres de las características 
dv.get_feature_names_out()
```

NORMALIZACION DE VARIABLES NUMERICAS

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(X_train)  # Ajustar a los datos de entrenamiento
X_train_std = sc.transform(X_train)  # Normalización del conjunto de entrenamiento
X_test_std = sc.transform(X_test)  # Normalización del conjunto de prueba
```

```python
# Variable objetivo 
y_train = df_train_full['species'] 
y_test = df_test['species']
```

------------ENTRENAMIENTO DE LOS MODELOS-----------

ENTRENAMIENTO REGRESION LOGISTICA

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=1, max_iter=500)
lr.fit(X_train_std, y_train)
```

ENTRENAMIENTO SVM

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear', probability=True, random_state=1)
svm.fit(X_train_std, y_train)
```

ENTRENAMIENTO DECISSION TREES

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=1)
dt.fit(X_train_std, y_train)
```

ENTRENAMIENTO KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_std, y_train)
```

SERIALIZACION DE LOS MODELOS

```python
import pickle

with open('../modelos/lr.pck', 'wb') as f:
    pickle.dump((sc, lr), f)

with open('../modelos/svm.pck', 'wb') as f:
    pickle.dump((sc, svm), f)

with open('../modelos/dt.pck', 'wb') as f:
    pickle.dump((sc, dt), f)

with open('../modelos/knn.pck', 'wb') as f:
    pickle.dump((sc, knn), f)
```
